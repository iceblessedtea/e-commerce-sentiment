```python
#distribution sentiment 
import matplotlib.pyplot as plt
import seaborn as sns

sentiment_count = df['sentiment'].value_counts()
sns.set_style("whitegrid")

fig, ax = plt.subplots(figsize=(6, 4))
ax = sns.barplot(x=sentiment_count.index, y=sentiment_count.values, palette="pastel")
plt.title("Distribusi Sentimen", fontsize=15, pad=20)
plt.xlabel("Sentimen", fontsize=12, labelpad=15)
plt.ylabel("Jumlah", fontsize=12, labelpad=15)

total = len(df['sentiment'])
for i, count in enumerate(sentiment_count):
    percentage = f'{100 * count / total:.2f}%'
    ax.text(i, count, percentage, ha='center', va='bottom', fontsize=12)

plt.show()
```

```python
from sklearn.utils import resample

# Pisahkan data berdasarkan kelas
df_pos = df[df['sentiment'] == 'Positif']
df_net = df[df['sentiment'] == 'Netral']
df_neg = df[df['sentiment'] == 'Negatif']

# Cari jumlah maksimum dari ketiga kelas
max_count = max(len(df_pos), len(df_net), len(df_neg))

# Oversample setiap kelas ke jumlah maksimum
df_pos_upsampled = resample(df_pos, replace=True, n_samples=max_count, random_state=42)
df_net_upsampled = resample(df_net, replace=True, n_samples=max_count, random_state=42)
df_neg_upsampled = resample(df_neg, replace=True, n_samples=max_count, random_state=42)

# Gabungkan kembali
df_balanced = pd.concat([df_pos_upsampled, df_net_upsampled, df_neg_upsampled]).sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
#distribution sentiment 
import matplotlib.pyplot as plt
import seaborn as sns

sentiment_count = df_balanced['sentiment'].value_counts()
sns.set_style("whitegrid")

fig, ax = plt.subplots(figsize=(6, 4))
ax = sns.barplot(x=sentiment_count.index, y=sentiment_count.values, palette="pastel")
plt.title("Distribusi Sentimen", fontsize=15, pad=20)
plt.xlabel("Sentimen", fontsize=12, labelpad=15)
plt.ylabel("Jumlah", fontsize=12, labelpad=15)

total = len(df['sentiment'])
for i, count in enumerate(sentiment_count):
    percentage = f'{100 * count / total:.2f}%'
    ax.text(i, count, percentage, ha='center', va='bottom', fontsize=12)

plt.show()
```
